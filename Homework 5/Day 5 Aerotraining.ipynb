```python
import numpy as np
import pandas as pd
```

```python
#Insert dataframe

PaxData = pd.read_csv('train.csv')
PaxData
```

```python
#Dataframe clean-up

PaxData.drop(columns=['PassengerId','Name','Ticket','Fare','Cabin','Embarked'],inplace=True)
PaxData
```

```python
#Dataframe N/A values clean-up

PaxData.dropna(inplace=True)
PaxData
```

```python
#Make dummy variables

PaxData = pd.get_dummies(PaxData)
PaxData
```

```python
#Decision Tree

x = PaxData.drop(columns=['Survived'])
y = PaxData['Survived']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()

tree_model.fit(x_train, y_train)
```

```python
tree_model.score(x_test,y_test)
```

```python
sample_PaxData = x.iloc[540].tolist()
sample_PaxData
```

```python
tree_model.predict([sample_PaxData])
```

```python
#Input test dataframe

TestData = pd.read_csv('test.csv')
TestData
```

```python
#Test dataframe clean-up

TestData.dropna(inplace=True)
TestData.drop(columns=['PassengerId','Name','Ticket','Fare','Cabin','Embarked'], inplace=True)
TestData = pd.get_dummies(TestData)
TestData
```
